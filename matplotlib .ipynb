```python
# importing required libraries
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sb
from scipy.optimize import curve_fit
import pandas as pd
import numpy as np
```

### We have the min and max temperatures in a city In India for each months of the year. We would like to find a function to describe this and show it graphically, the dataset given below. Task: 
### 1. fitting it to the periodic function
### 2. plot the fit Data
### Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
### Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
# defining a periodic function to fit the data
def func(x, a, b, c,d):
    x=np.array(x)
    return a*np.sin(b*x + c) + d
```

```python
t_max = [39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25] # maximum temperatures
x = np.arange(0,len(t_max),1) # month
param1, param_cov1 = curve_fit(func, x, t_max) # fitting data to the defined function
```

```python
t_min = [21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]# minimum temperatures
x = np.arange(0,len(t_min),1)# month
param2, param_cov2 = curve_fit(func, x, t_min) # fitting data to the defined function
```

```python
param1 # parameters of function fitted on  t_max data 
```

```python
param2 # parameters of function fitted on  t_min data 
```

```python
# plotting the data with fitted functions
plt.plot(x, t_max, 'o', color ='red') 
plt.plot(x, func(x, *param1 ), color ='red')
plt.plot(x, t_min, 'o', color ='blue', label ="data") 
plt.plot(x, func(x, *param2 ), color ='blue') 
plt.xlabel('Month', fontsize = 13)
plt.ylabel('Temperature(deg. Celsius)', fontsize = 13)
plt.show() 
```

```python
# loading data from the given url
data = pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv')
```

```python
data.head()
```

```python
# creating series to plot the gender proportion
series = data['sex'].value_counts()
series.name = ''
```

```python
# plotting gender proportion
mpl.rcParams['font.size'] = 15
series.plot.pie(figsize = (7,7))
plt.title('Male/Female proportion')
plt.show()
```

```python
# plotting scatter plot of 'fare' and 'age' column with gender as the hue parameter
plt.figure(figsize = (10,8))
sb.scatterplot(x='age', y = 'fare', data = data, hue= 'sex' ,palette='deep')
plt.show()
```
