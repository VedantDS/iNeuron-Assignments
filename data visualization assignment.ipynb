###  In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
# importing required packages and libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# loading data
iris = load_iris()
X = iris.data
y = iris.target

fig = plt.figure(figsize=(10, 10))
ax = Axes3D(fig, elev=-150, azim=110)

# reducing the dimenions of the variables to 3 using PCA
pca = PCA(n_components=3)
pca.fit(X)
X = pca.transform(X)


# plotting 3-D scatter plot
sc = ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap="Spectral", edgecolor='k', s= 55,alpha = 0.7)

# setting legend for the corresponding species class
labelTups = [('Setosa', 0), ('Versicolour', 1), ('Virginica', 2)]
colors = [sc.cmap(sc.norm(i)) for i in [0,1,2]]
custom_lines = [plt.Line2D([],[], ls="", marker='.', 
                mec='k', mfc=c, mew=.1, ms=20) for c in colors]
ax.legend(custom_lines, [i[0] for i in labelTups], 
          loc='center left', bbox_to_anchor=(1.0, .5), fontsize = 15)

# setting axes and title name
ax.set_title("PCA with respect to flower species", fontsize = 15)
ax.set_xlabel("1st eigenvector", fontsize = 15)
ax.set_ylabel("2nd eigenvector", fontsize = 15)
ax.set_zlabel("3rd eigenvector", fontsize = 15)

plt.show()
```
